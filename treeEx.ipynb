```python
from rdkit import RDLogger
from rdkit import Chem
from MotifGraph import JTConstruct
```

```python
smiles = "C1=NC2=C(N1COCCO)N=C(NC2=O)N"   
mol = Chem.MolFromSmiles(smiles)
order, mol_tree,graph = JTConstruct(mol, 'brics').label_tree()
order
```

```python
from MotifGraph import tree_vis, vis_compare
tree = tree_vis(mol_tree, edge_labels=False, seed=42, show_clusters=True, figsize=(6,4))
```

```python
vis_compare(smiles, tree, figsize=(14, 6))
```
